```python
from bs4 import BeautifulSoup
import json

# html = '''
# <meta property="og:title" content="Maria Sharapova (@mariasharapova) ‚Ä¢ Instagram photos and videos">
# <meta property="og:url" content="https://www.instagram.com/mariasharapova/">
# <meta property="og:description" content="5M Followers, 707 Following, 1,993 Posts - See Instagram photos and videos from Maria Sharapova (@mariasharapova)">
# <meta content="5M Followers, 707 Following, 1,993 Posts - Maria Sharapova (@mariasharapova) on Instagram: &quot;Mama üêª&quot;" name="description">
# <link rel="canonical" href="https://www.instagram.com/mariasharapova/">
# '''
with open("test1.html", "r", encoding="utf-8") as file:
    html = file.read()

# Parse the HTML
soup = BeautifulSoup(html, 'html.parser')

# Extract meta tag with followers, following, posts
desc_meta = soup.find('meta', attrs={'property': 'og:description'}) or soup.find('meta', attrs={'name': 'description'})
desc_content = desc_meta['content'] if desc_meta else ''

# Extract username and name from og:title
title_meta = soup.find('meta', attrs={'property': 'og:title'})
title_content = title_meta['content'] if title_meta else ''

# Parse values
import re

match = re.search(r"([\d.,A-Z]+) Followers, ([\d.,A-Z]+) Following, ([\d.,A-Z]+) Posts", desc_content)
followers, following, posts = match.groups() if match else (None, None, None)

username_match = re.search(r"\(@([^)]+)\)", title_content)
username = username_match.group(1) if username_match else None

name_match = re.match(r"^(.*?) \(@", title_content)
name = name_match.group(1) if name_match else None

# Create final JSON
profile_info = {
    "name": name,
    "username": username,
    "followers": followers,
    "following": following,
    "posts": posts,
    "profile_url": soup.find('meta', attrs={'property': 'og:url'})['content']
}

# Print formatted JSON
print(json.dumps(profile_info, indent=2))
```

```python
from bs4 import BeautifulSoup

def extract_post_links(html_file_path, limit=7):
    with open(html_file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    post_links = []
    anchors = soup.find_all('a', href=True)

    for a in anchors:
        href = a['href']

        # Check if it's a post URL (like /p/POST_ID/)
        if '/p/' in href and href.startswith('/'):
            full_link = "https://www.instagram.com" + href

            # Avoid duplicates and profile or non-post links
            if full_link not in post_links:
                post_links.append(full_link)

        if len(post_links) == limit:
            break

    return post_links
```

```python
from bs4 import BeautifulSoup

def extract_instagram_caption(html_file):
    with open(html_file, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    # Find the h1 tag with all those classes
    caption_tag = soup.find("h1", class_="_ap3a _aaco _aacu _aacx _aad7 _aade")
    
    if caption_tag:
        return caption_tag.get_text(strip=True)
    return None

# Example usage
caption = extract_instagram_caption("test2.html")
print("Caption:", caption)
```

```python
from bs4 import BeautifulSoup
import json

def extract_image_info(html_file):
    with open(html_file, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    image_info = {}

    # Find the <img> tag inside <div class="_aagv">
    img_div = soup.find('div', class_='_aagv')
    if img_div:
        img_tag = img_div.find('img')
        if img_tag:
            image_info['image_url'] = img_tag.get('src')
            image_info['alt_text'] = img_tag.get('alt')

    return image_info

# Example usage
info = extract_image_info('test2.html')
print(json.dumps(info, indent=2, ensure_ascii=False))
```

```python
from bs4 import BeautifulSoup
import json

def extract_instagram_posts(html_file, top_k=5):
    with open(html_file, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    posts = []
    all_links = soup.find_all("a", class_="_a6hd")  # Each post <a> tag has this class

    for link in all_links[:top_k]:
        post = {}
        
        # Post URL
        href = link.get("href")
        if href:
            post["post_url"] = f"https://www.instagram.com{href}"

        # Image tag inside the post
        img_tag = link.find("img")
        if img_tag:
            post["image_url"] = img_tag.get("src")
            post["caption"] = img_tag.get("alt")

        posts.append(post)

    return posts


# Extract top 5 posts
results = extract_instagram_posts("test2.html")

# Print as JSON
print(json.dumps(results, indent=2, ensure_ascii=False))
```

```python
import re

def extract_instagram_handles(urls):
    instagram_handles = []

    for url in urls:
        # Match urls of the form instagram.com/handle/ (excluding /p/, /reel/, etc.)
        match = re.match(r"https?://(www\.)?instagram\.com/([a-zA-Z0-9_.]+)/?$", url)
        if match:
            handle = match.group(2)
            instagram_handles.append({
                "handle": handle,
                "url": url
            })

    return instagram_handles
```
